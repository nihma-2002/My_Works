```r
#Q2
Y=c(132,143,153,162,154,168,137,149,159,128,166)
x1=c(52,59,67,73,64,74,54,61,65,46,72)
x2=c(173,184,194,211,196,220,188,188,207,167,217)
model=lm(Y~x1+x2);model
names(model)
a=model
a

#b
residual=model$residuals
residual
var(residual)

#c
summary(a)
#97% of variance of Y is explained by x1 and x2

#d
par(mfrow=c(2,2))
plot(a)
mean(residual)
#The mean of the resiual is very close to 0. The assumptions are vaild

#Q3
pd=c(7,3,3,4,6,7,2,7,30,5,16,10,6,9,10,6,7,3,17,10,26,9,8,4)
dist=c(560,220,340,80,150,330,110,210,1460,605,888,215,462,448,776,200,132,36,770,140,810,450,635,150)
dt=c(17.08,11.5,12.03,15.28,14.15,18.11,8,18.23,79.24,21.5,40.33,21.06,20.15,24.36,29.53,15.35,19.11,9.5,35.1,18.3,52.32,19.15,20.23,11.15)
model_2=lm(dt~pd+dist)
b=model_2
b

#b,c
summary(b)
#All regression coefficents are significant

#d
plot(dt,pd)
plot(dt,dist)
boxplot(dt,pd)
boxplot(dt,dist)

#e
Residual=b$residuals
var(Residual)
mean(Residual)
par(mfrow=c(2,2))
plot(b)

#f
par(mfrow=c(1,1))
hist(Residual,probability=TRUE)
x=seq(-10,10,by=0.05)
y=dnorm(x,0,var(Residual))
lines(x,y)
#Its not normal distribution


#Q4
rate=c(4.58,2.86,3.02,2.29,1.61,6.87,3.85,6.12,3.29,5.88,4.2,3.85,1.99,4.22,2.76,2.55)
len=c(4.99,16.11,9.75,10.65,20.01,5.97,8.57,5.24,15.79,8.26,7.03,2.96,8.86,5.49,8.63,20.31)
adt=c(69,73,49,61,28,30,46,25,43,23,23,21,22,9,12,12)
trks=c(8,8,10,13,12,6,8,9,12,7,6,8,9,22,8,7)
sigs1=c(0.2004,0.0621,0.1026,0.0939,0.0499,2.0075,0.8167,0.5708,1.4533,1.3311,1.9922,0.3378,0.1129,0.3621,0.1159,1.0392)
lane=c(8,4,4,6,4,4,4,4,4,4,4,4,4,2,2,4)
acpt=c(4.6,4.4,4.7,3.8,2.2,24.8,11,18.5,7.5,8.2,5.4,5.4,3.2,8.9,12.4,7.8)
itg=c(1.2,1.43,1.54,0.94,0.65,0.34,0.47,0.38,0.95,0.12,0.29,0.34,0.68,0.18,0.14,0.05)
data=data.frame(rate,len,adt,trks,sigs1,lane,acpt,itg)
data

#a
install.packages("psych")
library(psych)
summary(data)
describe(data)

#b
model_3=lm(rate~.,data=data)
c=model_3
c

#c
summary(c)

#d
d=lm(rate~len+acpt)
d

#e
summary(d)

#f
Residuals=c$residuals
Residuals
mean(Residuals)
par(mfrow=c(2,2))
plot(c)
```
